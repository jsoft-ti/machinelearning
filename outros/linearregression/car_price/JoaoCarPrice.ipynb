```python
import pandas as pd
import numpy as np
```

```python
from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)
```

```python
def EDA (df, order):
    eda_df = {}
    eda_df['Amount_NaN'] = df.isnull().sum()
    eda_df['%_NaN'] = df.isnull().mean().round(2)
    eda_df['DType'] = df.dtypes
    eda_df['Amount_Data'] = df.count()
    
    if order == True:
        colunas = sorted(df.columns.tolist(), key=str.lower, reverse=False)
    else:
        colunas = df.columns.tolist()
    eda_df['Amount_Unique'] = filtered_result = list(map(lambda x: len(df[x].unique().tolist()), colunas))
    
    eda_df['Mean'] = np.round(df.mean(), 2)
    eda_df['Median'] = np.round(df.median(), 2)
    
    eda_df['Max'] = df.max()
    eda_df['Min'] = df.min()
    eda_df['STD'] = np.round(df.std(), 2)
    
    return pd.DataFrame(eda_df)
```

```python
# Análise inicial:

#Não há dados faltantes

#a coluna id é a identificação do pedido - não trás poder preditivo - podemos excluir essa coluna 
del df['id']
```

```python
# Segunda etapa da limpeza
# Temos colunas que são um identificador porém são no tipo inteiro, o que fazer?
# essas colunas são center_id e meal_id
```

```python
## excluir para os alunos

df2 = df.merge(df_comida, how = 'left', on='meal_id')
df2 = df2.merge(df_centro, how = 'left', on='center_id')

df.head()

hide_toggle()
```

```python
df2.head()
```

```python
## Não vamos mais precisar das colunas center_id e meal_id pois já temos todas as informações 
del df2['center_id']
del df2['meal_id']
```

```python
df2.head()
```

```python
## hora de escolher quais colunas fazem sentido manter para fazer uma estimativa de consumo

# checkout_price : é uma informação que não temos antes da compra ser fechada
################## temos essa informação para o passado mas não temos essa informação para o futuro

# base_price : podemos manter essa coluna, pois os preços são pré definidos

# homepage_featured : não sabemos por onde o consumidor irá fechar a compra
####################  nesse primeiro momento podemos excluir, porém para futuros trabalhos podemos utilizar essa 
#################### coluna

# emailer_for_promotion : podemos manter, pois saberemos os emails com promoção que serão enviados com antecedência


df3 = df2.drop(['checkout_price', 'homepage_featured'], axis = 1)
df3.head()
```

```python
# podemos selecionar as colunas pata fazermos o get_dummy

# colunas interessantes para o get_dummy(category, cusine, center_type)

# colunas para serem trabalhadas com frequency encoder 
```

```python
df4 = pd.get_dummies(df3, columns = ['category', 'cuisine', 'center_type'])
hide_toggle()
```

```python
df4.head()
```

```python
df4['city_code'] = df4.city_code.astype(str)
df4['region_code'] = df4.region_code.astype(str)
```

```python
## Vamos fazer a seleção separar o dados de treino e teste
treino_df = df4.loc[df4['week'] <= 10]

y_treino = treino_df['num_orders']
del treino_df['num_orders']
del treino_df['week']
```

```python
teste_df = df4.loc[(df4['week'] > 11) & (df4['week'] <= 12)]

y_teste = teste_df['num_orders']
del teste_df['num_orders']
del teste_df['week']
```

```python
## algumas libs para trabalhar com variáveis categóricas 
## https://contrib.scikit-learn.org/category_encoders/
## https://feature-engine.readthedocs.io/en/latest/index.html
```

```python
from feature_engine import categorical_encoders as ce

encoder = ce.CountFrequencyCategoricalEncoder(encoding_method='frequency',
                         variables=['city_code', 'region_code'])

# treinar o encoder
encoder.fit(treino_df)

# transformar os dados
train_t = encoder.transform(treino_df)
test_t = encoder.transform(teste_df)
```

```python
# Padronizar os dados
from sklearn import preprocessing

#definir o scaler
scaler = preprocessing.RobustScaler().fit(train_t.values)

#transformar o dado
x_train_norm = scaler.transform(train_t.values)
x_test_norm = scaler.transform(test_t.values)
```

```python
# regressão linear
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x_train_norm, y_treino)

reg.score(x_train_norm, y_treino)
```

```python
# calcular a métrica da estimativa no conjunto teste
from sklearn.metrics import r2_score

y_pred = reg.predict(x_test_norm)
r2_score(y_teste, y_pred)
```

```python
# importância das features
importance_df = pd.DataFrame()
importance_df['colunas'] = teste_df.columns.tolist()
importance_df['importancia'] = list(reg.coef_)
importance_df['importancia_abs'] = np.abs(list(reg.coef_))

importance_df = importance_df.sort_values(by='importancia_abs', ascending=False)
del importance_df['importancia_abs']

importance_df
```

```python
import matplotlib.pyplot as plt

min_index = 0
max_index = 100

fig = plt.figure()

ax1 = fig.add_subplot(111)

ax1.plot(test_t.index[min_index:max_index], y_teste[min_index:max_index], label='Real')
ax1.scatter(test_t.index[min_index:max_index], y_pred[min_index:max_index], s=10, c='r', marker="o", \
            label='Projeção')
plt.legend(loc='upper center')
plt.show()
```

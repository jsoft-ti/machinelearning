```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
#https://archive.ics.uci.edu/ml/datasets/Appliances+energy+prediction#
```

```python
def EDALUCIO (df, order):
    eda_df = {}
    eda_df['Amount_NaN'] = df.isnull().sum()
    eda_df['%_NaN'] = df.isnull().mean().round(2)
    eda_df['DType'] = df.dtypes
    eda_df['Amount_Data'] = df.count()
    
    if order == True:
        colunas = sorted(df.columns.tolist(), key=str.lower, reverse=False)
    else:
        colunas = df.columns.tolist()
    eda_df['Amount_Unique'] = filtered_result = list(map(lambda x: len(df[x].unique().tolist()), colunas))
    
    eda_df['Mean'] = np.round(df.mean(), 2)
    eda_df['Median'] = np.round(df.median(), 2)
    
    eda_df['Max'] = df.max()
    eda_df['Min'] = df.min()
    eda_df['STD'] = np.round(df.std(), 2)
    
    return pd.DataFrame(eda_df)
```

```python

df = pd.read_csv('data/energydata_complete.csv')
df.head()
EDALUCIO(df, True)
```

Where indicated, hourly data (then interpolated) from the nearest airport weather station (Chievres Airport, Belgium) was downloaded from a public data set from Reliable Prognosis, rp5.ru. Permission was obtained from Reliable Prognosis for the distribution of the 4.5 months of weather data.

```python
justDate    = df["date"].str.split(" ").str[0]
justTime    = df["date"].str.split(" ").str[1]
splitDate   = justDate.str.split("-")
splitTime   = justTime.str.split(":")
df["hour"]  = splitTime.str[0].astype(int)
df["min"]   = splitTime.str[1].astype(int)
df["day"]   = splitDate.str[2].astype(int)
df["month"] = splitDate.str[1].astype(int)
df["year"]  = splitDate.str[0].astype(int)


df[["year","month","day","hour", "min"]]
del df["date"]
```

```python
EDALUCIO(df, True)
```

```python
treino_df = df.loc[df['month'] <= 4]
y_treino_rv1 = treino_df['rv1']
y_treino_rv2 = treino_df['rv2']
##Perguntar para o Lúcio pq ele retirou a semana do exercício nesta parte
del treino_df['rv1']
del treino_df['rv2']
```

```python
teste_df = df.loc[(df['month'] == 5)]
y_teste_rv1 = teste_df['rv1']
y_teste_rv2 = teste_df['rv2']
del teste_df['rv1']
del teste_df['rv2']
```

```python
#definir o scaler
scaler = preprocessing.StandardScaler().fit(treino_df.values)

#transformar o dado
x_train_norm = scaler.transform(treino_df.values)
x_test_norm = scaler.transform(teste_df.values)
```

```python
reg = LinearRegression().fit(x_train_norm, y_treino_rv1)

reg.score(x_train_norm, y_treino_rv1)
```

```python
y_pred = reg.predict(x_test_norm)
r2_score(y_teste_rv1, y_pred)
```
